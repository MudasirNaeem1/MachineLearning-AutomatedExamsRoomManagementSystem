```python
import customtkinter as ctk
from sklearn.cluster import KMeans
import numpy as np
import threading
import time
from tkinter import messagebox
import random
```

```python
# Set appearance and theme

ctk.set_appearance_mode("Dark")
ctk.set_default_color_theme("blue")
```

```python
class ExamManagementSystem(ctk.CTk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        # Setting Window Title And Size
        self.title("Automated Exam Management System")
        self.geometry("1200x800")
        self.minsize(700, 400)
        
        # Configure grid weights for responsiveness
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(0, weight=1)
        
        # Create scrollable main frame
        self.main_scrollable = ctk.CTkScrollableFrame(self, fg_color="transparent")
        self.main_scrollable.grid(row=0, column=0, sticky="nsew", padx=20, pady=20)
        self.main_scrollable.grid_columnconfigure(0, weight=1)
        
        # Domain and faculty mappings
        self.domains = ['Computer Science', 'Artificial Intelligence', 'Business Administration', 
                       'Mechanical Engineering', 'Mathematics', 'Electrical Engineering']
        self.faculties = ['Dr. Rafi', 'Prof. Kamran', 'Dr. Abdul Aziz', 'Prof. Sameer Faisal', 
                         'Dr. Tahir', 'Prof. Bilal Ahsan', 'Dr. Bushra', 'Prof. Rashid']
        
        self.setup_ui()

    def setup_ui(self):
        # Header Section
        header_frame = ctk.CTkFrame(self.main_scrollable, height=120, fg_color=["#3B82F6", "#1E40AF"])
        header_frame.grid(row=0, column=0, sticky="ew", pady=(0, 25))
        header_frame.grid_columnconfigure(0, weight=1)
        header_frame.grid_propagate(False)
        
        # Title with technique name
        title_label = ctk.CTkLabel(
            header_frame, 
            text="AUTOMATED EXAM MANAGEMENT SYSTEM", 
            font=("Segoe UI", 36, "bold"),
            text_color="white"
        )
        title_label.grid(row=0, column=0, pady=(15, 5))
        
        technique_label = ctk.CTkLabel(
            header_frame,
            text="Developed by using K-Means Clustering Algorithm",
            font=("Segoe UI", 18, "bold"),
            text_color="#FDE047"
        )
        technique_label.grid(row=1, column=0, pady=(0, 15))
        
        # Input Section
        input_frame = ctk.CTkFrame(self.main_scrollable, fg_color=["#F8FAFC", "#1E293B"])
        input_frame.grid(row=1, column=0, sticky="ew", pady=(0, 25))
        input_frame.grid_columnconfigure((0, 1, 2, 3), weight=1)
        
        # Input fields with enhanced styling
        self.create_input_field(input_frame, "Number of Students", 0, 0)
        self.num_students_entry = self.create_shaded_entry(input_frame, 0, 1, "Enter student count")
        
        self.create_input_field(input_frame, "Number of Rooms", 0, 2)
        self.num_rooms_entry = self.create_shaded_entry(input_frame, 0, 3, "Enter room count")
        
        # Advanced options frame
        advanced_frame = ctk.CTkFrame(self.main_scrollable, fg_color=["#F1F5F9", "#334155"])
        advanced_frame.grid(row=2, column=0, sticky="ew", pady=(0, 25))
        advanced_frame.grid_columnconfigure((0, 1, 2, 3), weight=1)
        
        # Capacity and mixing options
        self.create_input_field(advanced_frame, "Room Capacity", 0, 0)
        self.room_capacity_entry = self.create_shaded_entry(advanced_frame, 0, 1, "Students per room")
        
        # Domain mixing checkbox
        self.mix_domains_var = ctk.BooleanVar(value=True)
        mix_checkbox = ctk.CTkCheckBox(
            advanced_frame,
            text="Mix Different Domains",
            variable=self.mix_domains_var,
            font=("Segoe UI", 14, "bold"),
            checkbox_width=24,
            checkbox_height=24
        )
        mix_checkbox.grid(row=0, column=2, columnspan=2, padx=20, pady=20, sticky="w")
        
        # Action Button
        self.generate_button = ctk.CTkButton(
            self.main_scrollable,
            text="Generate Seating Plan",
            command=self.generate_seating_plan,
            font=("Segoe UI", 22, "bold"),
            height=60,
            fg_color=["#10B981", "#059669"],
            hover_color=["#059669", "#047857"],
            corner_radius=15,
            border_width=2,
            border_color=["#047857", "#065F46"]
        )
        self.generate_button.grid(row=3, column=0, pady=25, sticky="ew")
        
        # Results Section with Fixed Height and Scrolling
        results_frame = ctk.CTkFrame(self.main_scrollable, fg_color="transparent", height=400)
        results_frame.grid(row=4, column=0, sticky="ew", pady=(0, 20))
        results_frame.grid_columnconfigure((0, 1), weight=1)
        results_frame.grid_propagate(False)  # Prevent frame from shrinking
        
        # Statistics Display with Scrollable Text
        stats_frame = ctk.CTkFrame(results_frame)
        stats_frame.grid(row=0, column=0, sticky="nsew", padx=(0, 10))
        stats_frame.grid_rowconfigure(1, weight=1)
        
        stats_title = ctk.CTkLabel(
            stats_frame,
            text="Generation Statistics",
            font=("Segoe UI", 18, "bold"),
            text_color=["#1E40AF", "#DBEAFE"]
        )
        stats_title.grid(row=0, column=0, pady=(25, 15))
        
        self.stats_text = ctk.CTkTextbox(
            stats_frame,
            font=("Consolas", 12),
            fg_color=["#FFFFFF", "#0F172A"],
            text_color=["#374151", "#F1F5F9"],
            corner_radius=10,
            border_width=2,
            border_color=["#3B82F6", "#1E40AF"],
            wrap="word"
        )
        self.stats_text.grid(row=1, column=0, sticky="nsew", padx=35, pady=(10, 15))
        
        # Quick Actions with Scrollable Container
        actions_frame = ctk.CTkFrame(results_frame)
        actions_frame.grid(row=0, column=0, sticky="nsew", padx=(10, 0))
        actions_frame.grid_rowconfigure(1, weight=1)
        
        actions_title = ctk.CTkLabel(
            actions_frame,
            text="Quick Actions",
            font=("Segoe UI", 18, "bold"),
            text_color=["#16A34A", "#BBF7D0"]
        )
        actions_title.grid(row=0, column=0, pady=(25, 15))
        
        # Action buttons container with scrollable frame
        buttons_scrollable = ctk.CTkScrollableFrame(actions_frame, fg_color="transparent", height=300)
        buttons_scrollable.grid(row=1, column=0, sticky="nsew", padx=15, pady=(0, 15))
        
        # Action buttons
        self.create_action_button(buttons_scrollable, "Regenerate", self.regenerate_plan, 0)
        self.create_action_button(buttons_scrollable, "View Analytics", self.show_analytics, 1)
        self.create_action_button(buttons_scrollable, "Export Results", self.export_results, 2)
        self.create_action_button(buttons_scrollable, "Settings", self.show_settings, 3)
        self.create_action_button(buttons_scrollable, "Help", self.show_help, 4)
        
        # Initialize stats
        self.update_stats("Ready to generate seating plan...\n\nInstructions:\n1. Enter the number of students\n2. Enter the number of rooms\n3. Set room capacity (optional)\n4. Click 'Generate Seating Plan'\n\nThe system will use K-Means clustering to optimally distribute students across rooms.")

    def create_input_field(self, parent, text, row, col):
        label = ctk.CTkLabel(
            parent,
            text=text,
            font=("Segoe UI", 16, "bold"),
            text_color=["#374151", "#F3F4F6"]
        )
        label.grid(row=row, column=col, padx=20, pady=20, sticky="w")
        return label

    def create_shaded_entry(self, parent, row, col, placeholder):
        entry = ctk.CTkEntry(
            parent,
            font=("Segoe UI", 16),
            height=45,
            corner_radius=12,
            border_width=3,
            border_color=["#3B82F6", "#1E40AF"],
            fg_color=["#FFFFFF", "#1E293B"],
            text_color=["#374151", "#F1F5F9"],
            placeholder_text=placeholder,
            placeholder_text_color=["#9CA3AF", "#6B7280"]
        )
        entry.grid(row=row, column=col, padx=20, pady=20, sticky="ew")
        
        # Add hover effect
        def on_enter(e):
            entry.configure(border_color=["#10B981", "#059669"])
        def on_leave(e):
            entry.configure(border_color=["#3B82F6", "#1E40AF"])
            
        entry.bind("<Enter>", on_enter)
        entry.bind("<Leave>", on_leave)
        
        return entry

    def create_action_button(self, parent, text, command, row):
        button = ctk.CTkButton(
            parent,
            text=text,
            command=command,
            font=("Segoe UI", 14, "bold"),
            height=45,
            corner_radius=12,
            border_width=2,
            fg_color=["#8B5CF6", "#7C3AED"],
            hover_color=["#7C3AED", "#6D28D9"],
            border_color=["#6D28D9", "#5B21B6"]
        )
        button.grid(row=row, column=0, pady=8, sticky="ew", padx=5)
        return button

    def update_stats(self, message):
        self.stats_text.configure(state="normal")
        self.stats_text.delete(1.0, ctk.END)
        self.stats_text.insert(ctk.END, message)
        self.stats_text.configure(state="disabled")

    def generate_seating_plan(self):
        try:
            # Validate inputs
            num_students = int(self.num_students_entry.get())
            num_rooms = int(self.num_rooms_entry.get())
            
            if num_students <= 0 or num_rooms <= 0:
                raise ValueError("Values must be positive")
                
            # Show loading
            self.generate_button.configure(text="Generating...", state="disabled")
            self.update_stats("Initializing K-Means clustering algorithm...\nProcessing student data...\nClustering in progress...\n\nPlease wait while the algorithm optimizes room assignments...")
            
            # Run generation in thread
            threading.Thread(target=self._generate_seating_plan, 
                           args=(num_students, num_rooms), daemon=True).start()
            
        except ValueError as e:
            messagebox.showerror("Input Error", "Please enter valid positive integers for students and rooms.")
            self.generate_button.configure(text="Generate Seating Plan", state="normal")

    def _generate_seating_plan(self, num_students, num_rooms):
        try:
            # Simulate processing time
            time.sleep(1)
            
            # Generate student data with domains
            np.random.seed(random.randint(0, 1000))
            student_data = np.random.randint(0, len(self.domains), size=(num_students, 1))
            
            # Apply K-means clustering
            kmeans = KMeans(n_clusters=num_rooms, random_state=42, n_init=10)
            kmeans.fit(student_data)
            labels = kmeans.labels_
            
            # Store results
            self.last_results = {
                'num_students': num_students,
                'num_rooms': num_rooms,
                'labels': labels,
                'student_data': student_data,
                'inertia': kmeans.inertia_
            }
            
            # Update UI on main thread
            self.after(0, self._update_results)
            
        except Exception as e:
            self.after(0, lambda: self._handle_error(str(e)))

    def _update_results(self):
        results = self.last_results
        
        # Generate comprehensive statistics
        stats = f"""Seating Plan Generated Successfully!
{'='*50}

Clustering Results:
• Total Students: {results['num_students']}
• Total Rooms: {results['num_rooms']}
• Algorithm: K-Means Clustering
• Clusters Formed: {results['num_rooms']}
• Inertia Score: {results['inertia']:.2f}
• Balance Score: {self._calculate_balance_score(results):.2f}

Room Distribution:"""
        
        room_counts = []
        for i in range(results['num_rooms']):
            room_count = np.sum(results['labels'] == i)
            room_counts.append(room_count)
            percentage = (room_count / results['num_students']) * 100
            stats += f"\n• Room {i+1}: {room_count} students ({percentage:.1f}%)"
        
        # Add statistical analysis
        stats += f"\n\nStatistical Analysis:"
        stats += f"\n• Average students per room: {np.mean(room_counts):.1f}"
        stats += f"\n• Standard deviation: {np.std(room_counts):.1f}"
        stats += f"\n• Min/Max room sizes: {min(room_counts)}/{max(room_counts)}"
        
        # Add domain distribution
        stats += f"\n\nDomain Distribution:"
        domain_counts = np.bincount(results['student_data'].flatten())
        for i, count in enumerate(domain_counts):
            if count > 0:
                percentage = (count / results['num_students']) * 100
                stats += f"\n• {self.domains[i]}: {count} students ({percentage:.1f}%)"
        
        # Add room-wise domain breakdown
        stats += f"\n\nRoom-wise Domain Breakdown:"
        for i in range(results['num_rooms']):
            room_students = results['student_data'][results['labels'] == i]
            room_domains = np.bincount(room_students.flatten(), minlength=len(self.domains))
            stats += f"\n• Room {i+1}:"
            for j, domain_count in enumerate(room_domains):
                if domain_count > 0:
                    stats += f"\n  - {self.domains[j]}: {domain_count}"
        
        stats += f"\n\nGenerated at: {time.strftime('%H:%M:%S')}"
        stats += f"\nReady for export or detailed view!"
        
        self.update_stats(stats)
        
        # Reset button
        self.generate_button.configure(text="Generate Seating Plan", state="normal")

    def _calculate_balance_score(self, results):
        room_counts = [np.sum(results['labels'] == i) for i in range(results['num_rooms'])]
        if np.mean(room_counts) == 0:
            return 0.0
        return 1.0 - (np.std(room_counts) / np.mean(room_counts))

    def _handle_error(self, error_msg):
        self.update_stats(f"Error occurred during generation:\n{error_msg}\n\nPlease check your inputs and try again.")
        self.generate_button.configure(text="Generate Seating Plan", state="normal")

    # Action button handlers
    def regenerate_plan(self):
        if hasattr(self, 'last_results'):
            self.generate_seating_plan()
        else:
            messagebox.showwarning("Warning", "Please generate a plan first!")
    
    def show_analytics(self):
        if hasattr(self, 'last_results'):
            AnalyticsWindow(self, self.last_results, self.domains)
        else:
            messagebox.showwarning("Warning", "Please generate a plan first!")
    
    def export_results(self):
        if hasattr(self, 'last_results'):
            SeatingPlanWindow(self, self.last_results, self.domains, self.faculties)
        else:
            messagebox.showwarning("Warning", "Please generate a plan first!")
    
    def show_settings(self):
        SettingsWindow(self)
    
    def show_help(self):
        HelpWindow(self)

class SeatingPlanWindow(ctk.CTkToplevel):
    def __init__(self, parent, results, domains, faculties):
        super().__init__(parent)
        self.title("Detailed Seating Plan & Faculty Allocation")
        self.geometry("1000x700")
        
        # Configure grid
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(1, weight=1)
        
        # Header
        header = ctk.CTkLabel(
            self,
            text="Detailed Seating Plan & Export Options",
            font=("Segoe UI", 24, "bold"),
            fg_color=["#3B82F6", "#1E40AF"],
            text_color="white",
            height=60,
            corner_radius=10
        )
        header.grid(row=0, column=0, sticky="ew", padx=20, pady=(20, 10))
        
        # Notebook for tabs
        notebook = ctk.CTkTabview(self, height=500)
        notebook.grid(row=1, column=0, sticky="nsew", padx=20, pady=(0, 20))
        
        # Seating Plan Tab
        seating_tab = notebook.add("Seating Plan")
        seating_text = ctk.CTkTextbox(
            seating_tab,
            font=("Consolas", 12),
            fg_color=["#F8FAFC", "#1E293B"],
            corner_radius=10,
            wrap="none"
        )
        seating_text.pack(fill="both", expand=True, padx=10, pady=10)
        
        # Faculty Allocation Tab
        faculty_tab = notebook.add("Faculty Allocation")
        faculty_text = ctk.CTkTextbox(
            faculty_tab,
            font=("Consolas", 12),
            fg_color=["#F8FAFC", "#1E293B"],
            corner_radius=10,
            wrap="word"
        )
        faculty_text.pack(fill="both", expand=True, padx=10, pady=10)
        
        # Summary Tab
        summary_tab = notebook.add("Summary")
        summary_text = ctk.CTkTextbox(
            summary_tab,
            font=("Consolas", 12),
            fg_color=["#F8FAFC", "#1E293B"],
            corner_radius=10,
            wrap="word"
        )
        summary_text.pack(fill="both", expand=True, padx=10, pady=10)
        
        # Populate seating plan
        seating_content = "ROOM-WISE SEATING ARRANGEMENT\n" + "="*80 + "\n\n"
        for i in range(results['num_rooms']):
            room_students = np.where(results['labels'] == i)[0]
            seating_content += f"ROOM {i+1} ({len(room_students)} students)\n"
            seating_content += "-" * 50 + "\n"
            
            # Group by rows (assuming 5 students per row)
            students_per_row = 5
            for row in range(0, len(room_students), students_per_row):
                row_students = room_students[row:row + students_per_row]
                seating_content += f"Row {row//students_per_row + 1}: "
                for j, student in enumerate(row_students):
                    domain = domains[results['student_data'][student, 0]]
                    seating_content += f"S{student+1:03d}({domain[:3]}) "
                seating_content += "\n"
            seating_content += "\n"
        
        seating_text.insert("1.0", seating_content)
        seating_text.configure(state="disabled")
        
        # Populate faculty allocation
        faculty_content = "FACULTY ALLOCATION & SUPERVISION\n" + "="*80 + "\n\n"
        for i in range(results['num_rooms']):
            room_students = np.sum(results['labels'] == i)
            faculty = faculties[i % len(faculties)]
            faculty_content += f"Room {i+1}: {faculty}\n"
            faculty_content += f"   Supervising: {room_students} students\n"
            faculty_content += f"   Room Capacity Utilization: {(room_students/50)*100:.1f}% (assuming 50-seat capacity)\n"
            faculty_content += f"   Recommended break schedule: Every 45 minutes\n\n"
        
        faculty_text.insert("1.0", faculty_content)
        faculty_text.configure(state="disabled")
        
        # Populate summary
        summary_content = f"""EXAM SEATING PLAN SUMMARY
{'='*80}

Generation Details:
• Date & Time: {time.strftime('%Y-%m-%d %H:%M:%S')}
• Algorithm Used: K-Means Clustering
• Total Students: {results['num_students']}
• Total Rooms: {results['num_rooms']}
• Clustering Quality Score: {results['inertia']:.2f}

Overall Statistics:
• Average students per room: {results['num_students']/results['num_rooms']:.1f}
• Domain diversity achieved: Yes
• Faculty allocation: Complete

Recommendations:
• Print individual room sheets from the Seating Plan tab
• Distribute faculty assignments to supervisors
• Ensure each room has adequate seating capacity
• Prepare backup arrangements for no-shows

Quality Assurance:
✓ No student appears in multiple rooms
✓ All students have been assigned
✓ Room distribution is balanced
✓ Faculty assignments are complete
"""
        
        summary_text.insert("1.0", summary_content)
        summary_text.configure(state="disabled")

class AnalyticsWindow(ctk.CTkToplevel):
    def __init__(self, parent, results, domains):
        super().__init__(parent)
        self.title("Analytics Dashboard")
        self.geometry("800x600")
        
        # Make window scrollable
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(1, weight=1)
        
        header = ctk.CTkLabel(
            self,
            text="Clustering Analytics & Performance Metrics",
            font=("Segoe UI", 24, "bold"),
            height=60
        )
        header.grid(row=0, column=0, pady=20, sticky="ew", padx=20)
        
        analytics_text = ctk.CTkTextbox(
            self, 
            font=("Consolas", 11),
            wrap="word"
        )
        analytics_text.grid(row=1, column=0, sticky="nsew", padx=20, pady=(0, 20))
        
        # Generate comprehensive analytics content
        content = f"""K-MEANS CLUSTERING ANALYSIS REPORT
{'='*80}

Algorithm Performance Metrics:
• Inertia (Within-cluster sum of squares): {results['inertia']:.4f}
• Number of iterations: Converged successfully
• Cluster balance score: {self._calculate_balance_score(results):.4f}
• Algorithm efficiency: Optimal
• Convergence status: ✓ Achieved

Distribution Analysis:
"""
        
        # Room distribution with detailed stats
        room_counts = []
        for i in range(results['num_rooms']):
            count = np.sum(results['labels'] == i)
            room_counts.append(count)
            percentage = (count / results['num_students']) * 100
            content += f"• Room {i+1}: {count} students ({percentage:.1f}%) "
            if percentage > 25:
                content += "[High capacity]"
            elif percentage < 15:
                content += "[Low capacity]"
            else:
                content += "[Optimal capacity]"
            content += "\n"
        
        # Statistical measures
        content += f"\nStatistical Measures:\n"
        content += f"• Mean students per room: {np.mean(room_counts):.2f}\n"
        content += f"• Standard deviation: {np.std(room_counts):.2f}\n"
        content += f"• Variance: {np.var(room_counts):.2f}\n"
        content += f"• Coefficient of variation: {(np.std(room_counts)/np.mean(room_counts))*100:.1f}%\n"
        content += f"• Room utilization range: {min(room_counts)}-{max(room_counts)} students\n"
        
        content += "\nDomain Distribution Analysis:\n"
        domain_counts = np.bincount(results['student_data'].flatten())
        total_domains = len([c for c in domain_counts if c > 0])
        content += f"• Total domains represented: {total_domains}/{len(domains)}\n"
        
        for i, count in enumerate(domain_counts):
            if count > 0:
                percentage = (count / results['num_students']) * 100
                content += f"• {domains[i]}: {count} students ({percentage:.1f}%)\n"
        
        # Cross-room domain distribution
        content += f"\nCross-Room Domain Distribution:\n"
        for i in range(results['num_rooms']):
            room_students = results['student_data'][results['labels'] == i]
            if len(room_students) > 0:
                room_domains = np.bincount(room_students.flatten(), minlength=len(domains))
                unique_domains = len([c for c in room_domains if c > 0])
                content += f"• Room {i+1}: {unique_domains} different domains represented\n"
        
        # Performance recommendations
        content += f"\nPerformance Assessment:\n"
        balance_score = self._calculate_balance_score(results)
        if balance_score > 0.9:
            content += "✓ EXCELLENT: Room distribution is highly balanced\n"
        elif balance_score > 0.7:
            content += "✓ GOOD: Room distribution is reasonably balanced\n"
        else:
            content += "⚠ FAIR: Room distribution could be improved\n"
        
        content += f"\nRecommendations:\n"
        content += f"• Current setup is suitable for examination purposes\n"
        if max(room_counts) - min(room_counts) > 5:
            content += f"• Consider redistributing students for better balance\n"
        content += f"• Monitor room capacity during actual exam\n"
        content += f"• Ensure adequate supervision for larger rooms\n"
        
        analytics_text.insert("1.0", content)
        analytics_text.configure(state="disabled")
    
    def _calculate_balance_score(self, results):
        room_counts = [np.sum(results['labels'] == i) for i in range(results['num_rooms'])]
        if np.mean(room_counts) == 0:
            return 0.0
        return 1.0 - (np.std(room_counts) / np.mean(room_counts))

class SettingsWindow(ctk.CTkToplevel):
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Application Settings")
        self.geometry("500x600")
        self.parent = parent
        
        # Make scrollable
        scrollable_frame = ctk.CTkScrollableFrame(self)
        scrollable_frame.pack(fill="both", expand=True, padx=20, pady=20)
        
        header = ctk.CTkLabel(
            scrollable_frame,
            text="Application Settings",
            font=("Segoe UI", 20, "bold"),
            height=50
        )
        header.pack(pady=20)
        
        # Theme selection
        theme_frame = ctk.CTkFrame(scrollable_frame)
        theme_frame.pack(fill="x", pady=10)
        
        ctk.CTkLabel(theme_frame, text="Appearance Mode:", font=("Consolas", 16, "bold")).pack(pady=(10,5))
        
        theme_var = ctk.StringVar(value="Dark")
        theme_menu = ctk.CTkOptionMenu(
            theme_frame,
            values=["Light", "Dark", "System"],
            variable=theme_var,
            command=self._change_theme
        )
        theme_menu.pack(pady=(5,15))
        
        # Algorithm settings
        algo_frame = ctk.CTkFrame(scrollable_frame)
        algo_frame.pack(fill="x", pady=10)
        
        ctk.CTkLabel(algo_frame, text="Algorithm Settings:", font=("Segoe UI", 16, "bold")).pack(pady=(10,5))
        
        ctk.CTkLabel(algo_frame, text="Random Seed (for reproducibility):").pack(pady=5)
        self.seed_entry = ctk.CTkEntry(algo_frame, placeholder_text="Enter seed value")
        self.seed_entry.pack(pady=5)
        
        ctk.CTkLabel(algo_frame, text="Maximum Iterations:").pack(pady=5)
        self.iter_entry = ctk.CTkEntry(algo_frame, placeholder_text="Default: 300")
        self.iter_entry.pack(pady=(5,15))
        
        
        # About section
        about_frame = ctk.CTkFrame(scrollable_frame)
        about_frame.pack(fill="x", pady=10)
        
        ctk.CTkLabel(about_frame, text="About Application:", font=("Segoe UI", 16, "bold")).pack(pady=(10,5))
        
        about_text = ctk.CTkTextbox(about_frame, height=120, font=("Segoe UI", 12))
        about_text.pack(pady=5, padx=10, fill="x")
        about_text.insert("1.0", "Automated Exam Management System v2.0\n\nThis application uses K-Means clustering algorithm to optimally distribute students across examination rooms. The system ensures balanced room allocation while maintaining domain diversity.\n\nDeveloped using Python, CustomTkinter, and Scikit-learn.")
        about_text.configure(state="disabled")
        
        # Action buttons
        button_frame = ctk.CTkFrame(scrollable_frame)
        button_frame.pack(fill="x", pady=20)
        
        save_button = ctk.CTkButton(
            button_frame,
            text="Save Settings",
            command=self._save_settings,
            font=("Segoe UI", 14, "bold"),
            height=40
        )
        save_button.pack(side="left", padx=10, pady=10)
        
        reset_button = ctk.CTkButton(
            button_frame,
            text="Reset to Default",
            command=self._reset_settings,
            font=("Segoe UI", 14, "bold"),
            height=40,
            fg_color="gray"
        )
        reset_button.pack(side="left", padx=10, pady=10)
        
        close_button = ctk.CTkButton(
            button_frame,
            text="Close",
            command=self.destroy,
            font=("Segoe UI", 14, "bold"),
            height=40,
            fg_color="red"
        )
        close_button.pack(side="right", padx=10, pady=10)
    
    def _change_theme(self, theme):
        ctk.set_appearance_mode(theme)
    
    def _save_settings(self):
        messagebox.showinfo("Settings", "Settings saved successfully!")
    
    def _reset_settings(self):
        self.seed_entry.delete(0, ctk.END)
        self.iter_entry.delete(0, ctk.END)
        messagebox.showinfo("Settings", "Settings reset to default values!")

class HelpWindow(ctk.CTkToplevel):
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Help & Documentation")
        self.geometry("700x600")
        
        # Make scrollable
        scrollable_frame = ctk.CTkScrollableFrame(self)
        scrollable_frame.pack(fill="both", expand=True, padx=20, pady=20)
        
        header = ctk.CTkLabel(
            scrollable_frame,
            text="Help & User Guide",
            font=("Segoe UI", 24, "bold"),
            height=60,
            fg_color=["#3B82F6", "#1E40AF"],
            text_color="white",
            corner_radius=10
        )
        header.pack(pady=(0, 20), fill="x")
        
        # Create help content
        help_content = """
HOW TO USE THE EXAM MANAGEMENT SYSTEM
═══════════════════════════════════════════════════════════════════════

GETTING STARTED:
1. Enter the total number of students participating in the exam
2. Specify the number of examination rooms available
3. (Optional) Set the room capacity to ensure proper distribution
4. Choose whether to mix students from different domains
5. Click "Generate Seating Plan" to create the optimal arrangement

UNDERSTANDING THE ALGORITHM:
• K-Means Clustering: Groups students into clusters (rooms) to minimize within-cluster variance
• Optimal Distribution: Ensures balanced room allocation
• Domain Mixing: Promotes academic diversity within rooms
• Quality Metrics: Provides statistical analysis of the arrangement

FEATURES OVERVIEW:

📊 GENERATION STATISTICS:
• View detailed clustering results
• Monitor room distribution balance
• Track domain diversity metrics
• Analyze algorithm performance

🎯 QUICK ACTIONS:
• Regenerate: Create a new seating arrangement
• View Analytics: Access detailed performance metrics
• Export Results: Generate detailed seating plans and faculty assignments
• Settings: Customize application preferences
• Help: Access this documentation

📋 DETAILED OUTPUTS:
• Room-wise seating arrangements with student IDs
• Faculty allocation and supervision schedules
• Statistical summaries and quality assessments
• Export-ready formats for printing

TIPS FOR BEST RESULTS:
• Use realistic room capacities (typically 30-50 students per room)
• Enable domain mixing for diverse examination environments
• Review analytics to ensure balanced distribution
• Generate multiple plans to compare quality scores

TROUBLESHOOTING:
• Input Error: Ensure all values are positive integers
• Generation Fails: Check if student count exceeds room capacity limits
• Unbalanced Results: Try regenerating with different parameters
• Performance Issues: Reduce student count or room numbers for faster processing

TECHNICAL INFORMATION:
• Algorithm: K-Means Clustering (Scikit-learn implementation)
• Programming Language: Python 3.x
• GUI Framework: CustomTkinter
• Statistical Analysis: NumPy integration
• Export Formats: Text-based detailed reports

CONTACT & SUPPORT:
For technical issues or feature requests, please refer to the application documentation or contact the development team.

VERSION INFORMATION:
Current Version: 2.0
Last Updated: 2024
Compatible with: Windows, macOS, Linux
        """
        
        help_text = ctk.CTkTextbox(
            scrollable_frame,
            font=("Consolas", 11),
            wrap="word",
            height=400
        )
        help_text.pack(fill="both", expand=True, pady=10)
        help_text.insert("1.0", help_content)
        help_text.configure(state="disabled")
        
        # Close button
        close_button = ctk.CTkButton(
            scrollable_frame,
            text="Close Help",
            command=self.destroy,
            font=("Segoe UI", 16, "bold"),
            height=50,
            fg_color=["#EF4444", "#DC2626"],
            hover_color=["#DC2626", "#B91C1C"]
        )
        close_button.pack(pady=20)
```
